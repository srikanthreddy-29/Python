```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Create a plot figure 
x1 = np.linspace(0,10,100)
fig = plt.figure()
plt.plot(x1, np.sin(x1), '*')
plt.plot(x1, np.cos(x1), '--')
```

```python
# Create a plot figure 
x1 = np.linspace(0,10,100)
fig = plt.figure(figsize=(10,5))
plt.plot(x1,np.sin(x1),'--')
plt.plot(x1,np.cos(x1),'--')
```

```python
# Create a plot figure 
x1 = np.linspace(0,10,100)
fig = plt.figure()
plt.plot(x1, np.sin(x1), '*')
plt.plot(x1, np.cos(x1), '--')
plt.plot(x1,np.tan(x1),'*')
```

```python
#create a plot figure 
plt.figure()
# create the first of two panels and set current axis
plt.subplot(2,1,1)     #(rows,columns,panel)
plt.plot(x1,np.sin(x1))
# create the first of two panels and set current axis
plt.subplot(2,1,2)     #(rows,columns,panel)
plt.plot(x1,np.cos(x1))
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# call plot() method on the appropriate object 
ax[0].plot(x1, np.sin(x1),'bs')
ax[1].plot(x1,np.cos(x1),'g^')
```

```python
fig = plt.figure()
x2 = np.linspace(0,5,10)
y2 = x2**2
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x2,y2,'b--')
axes.set_xlabel('x2')
axes.set_ylabel('y2')
axes.set_title('title')
```

```python
# Adding a legend
x16 = np.arange(-2,10)
fig, ax = plt.subplots()

ax.plot(x16,x16*1.5)
ax.plot(x16,x16**2)
ax.plot(x16,x16/2.0)

ax.legend(["Normal","Fast","Slow"])
```

```python
x16 = np.arange(-2,10)
fig, ax = plt.subplots()

ax.plot(x16,x16*1.5,label="Normal")
ax.plot(x16,x16**2,label="Fast")
ax.plot(x16,x16/2.0,label="Slow")

ax.legend()
```

```python
#Control colours
plt.plot(x16,'r')
plt.plot(x16+2,'g')
plt.plot(x16-2,'b')
plt.show()
```

```python
# Control line Styles 
plt.plot(x16, '*',x16+1, '-.',x16+2, ':')
plt.show()
```
